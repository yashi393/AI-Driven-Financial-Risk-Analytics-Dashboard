```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/sampled_dataset_50000.csv")
```

```python
df_clean = df.copy()
```

```python
df_clean["date"] = pd.to_datetime(df_clean["date"])
```

```python
df_clean["date"].dtype
```

```python
df_clean["amount"].head()
```

```python
df_clean["amount"] = (
    df_clean["amount"]
    .str.replace("$", "", regex=False)
    .astype(float)
)
```

```python
df_clean["amount"].head()
```

```python
df_clean["error_flag"] = (
    ~df_clean["errors"].isna()
).astype("int8")
```

```python
df_clean["error_flag"].value_counts()
```

```python
df_clean["refund_flag"] = (
    df_clean["amount"] < 0
).astype("int8")
```

```python
df_clean["refund_flag"].value_counts()
```

```python
df_clean["year"] = df_clean["date"].dt.year

df_clean["month"] = df_clean["date"].dt.month

df_clean["month_name"] = df_clean["date"].dt.month_name()

df_clean["quarter"] = df_clean["date"].dt.quarter

df_clean["weekday"] = df_clean["date"].dt.day_name()

df_clean["hour"] = df_clean["date"].dt.hour
```

```python
df_clean["weekend_flag"] = (
    df_clean["weekday"]
    .isin(["Saturday", "Sunday"])
).astype("int8")
```

```python
df_clean["use_chip"].value_counts()
```

```python
df_clean.head()
```

```python
df_clean.to_csv(
    "../outputs/cleaned_transactions.csv",
    index=False
)
```
